```python
def load_data(fname = '/home/chun/Desktop/gpr_testing/auto_mpg/auto-mpg.csv'):
    X, Y = [], []
    with open(fname, 'r') as f:
        next(f)
        for line in f:
            d = line.split(',')
            try:
                X.append([float(a) for a in d[1:8]])
                Y.append(float(d[0]))
            except Exception as e:
#                 print(e)
                continue
                
    return np.array(X), np.array(Y)

X, Y = load_data()
# print(X, Y)
# print(list(X[0]))
```
